```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Entropia(x):
    if x == 0 or x == 1:
        return 0
    else:   
        return -(x*np.log(x)+(1-x)*np.log(1-x))
```

```python
x = np.linspace(0,1,101)
y = np.array([])
for i in range(len(x)):
    y = np.append(Entropia(x[i]),y)
```

```python
plt.plot(x,y)
```
