```python
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def probabilidad(p1,p2):
    return 0.5 * (p1* (1-p2) + p2* (1-p1)) + 0.25 * (p1* p2 + (1-p2) * (1-p1))
```

```python
p1 = np.linspace(0.1,0.9,100)
p2 = np.linspace(0.1,0.5,100)
X, Y = np.meshgrid(p1,p2)
Z = probabilidad(X,Y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection ='3d')
ax.plot_surface(Y,X,Z,cmap="coolwarm")
plt.show()
```

```python
max = 0
min = 1
for i in range(len(Z)):
    for j in range(len(Z)):
        if Z[i][j] >= max:
            max = Z[i][j]
        if Z[i][j] <= min:
            min = Z[i][j]
            
for i in range(len(p1)):
    for j in range(len(p2)):
        prob = probabilidad(p1[i],p2[j])
        if prob == max:
            punto1 = [p1[i],p2[j]]
        if prob == min:
            punto2 = [p1[i],p2[j]]
        
        
print("Máximo: "+str(max)+" en el punto " +str(punto1))
print("Mínimo: "+str(min)+" en el punto " +str(punto2))
```
