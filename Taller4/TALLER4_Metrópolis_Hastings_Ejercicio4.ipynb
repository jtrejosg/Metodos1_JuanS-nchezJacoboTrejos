```python
import numpy as np
import matplotlib.pyplot as plt

def cauchy_distribution(x):
  return 1 / (np.pi * (1 + x**2))

def metropolis(x0, N, function, delta= 0.5):
  
  x = np.zeros(N)
  x[0] = x0

  for i in range(0,N-1):
    
    proposal = x[i] + np.random.uniform(-delta, delta)

    if np.random.rand() < min(1, function(proposal)/function(x[i])):
      x[i+1] = proposal
    
    else:
      x[i+1] = x[i]

  return x

N = 1000
initparam = 0
samples = metropolis(initparam, N, cauchy_distribution)

plt.hist(samples, density=True, bins=50,label='Muestras')
xs =np.linspace(-10,10,100)
plt.plot(xs,cauchy_distribution(xs), 'r-', label='Distribución de Cauchy')
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.title('Muestreo de una distribución de Cauchy')
plt.legend()
plt.show()
```
