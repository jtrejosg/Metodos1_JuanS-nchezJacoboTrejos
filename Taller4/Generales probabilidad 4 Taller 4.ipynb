```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Cumpleaños(personas):
    resultado = 1
    for i in range(int(personas)):
        resultado *= (365-i)/365
    probabilidad = 1-resultado
    
    return probabilidad
```

```python
x = np.linspace(0,80,80)
y = np.array([])
for i in x:
    y = np.append(y,Cumpleaños(int(i)))
```

```python
plt.plot(x,y)
```
