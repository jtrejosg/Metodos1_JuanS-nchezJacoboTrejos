```python
import numpy as np
import matplotlib.pyplot as plt

def normal_distribution(x, mean = 2, std = 0.5):
  return (1 / (std * np.sqrt(2*np.pi))) * np.exp(-(x - mean)**2 / (2 * std**2))

def metropolis(x0, N, function, delta= 0.5):
  
  x = np.zeros(N)
  x[0] = x0

  for i in range(0,N-1):
    
    proposal = x[i] + np.random.uniform(-delta, delta)

    if np.random.rand() < min(1, function(proposal)/function(x[i])):
      x[i+1] = proposal
    
    else:
      x[i+1] = x[i]

  return x

N = 1000
initparam = 0
samples = metropolis(initparam, N, normal_distribution)

plt.hist(samples, density=True, bins=50,label='Muestras')
xs =np.linspace(0,4,100)
plt.plot(xs,normal_distribution(xs), 'r-', label='Distribución normal')
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.title('Muestreo de una distribución normal')
plt.legend()
plt.show()
```
