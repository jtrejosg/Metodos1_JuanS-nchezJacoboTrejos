```python
import numpy as np
import matplotlib.pyplot as plt
a = -1
b = 1
```

```python
def Function(x):
    return (x**3)/(np.exp(x)-1)
```

```python
def Function2(x):
    return ((x**3)*np.exp(x))/(np.exp(x)-1) #z
```

```python
def Integral(n):
    sumatoria = 0
    roots,weights = np.polynomial.laguerre.laggauss(n)
    
    for i in range(len(weights)):
        x = (a+b+roots[i]*(b-a))*0.5 #Cambio de base
        sumatoria += weights[i]*Function2(x)
        
    return sumatoria
```

```python
IntegralExacta = (np.pi**4)/15
print(Integral(3))
print(IntegralExacta)
```

```python
for i in range(2,11):
    error = Integral(i)/IntegralExacta
    plt.scatter(i,error)
```
