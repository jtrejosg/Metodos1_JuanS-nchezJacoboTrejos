```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
h = 0.001
x = np.linspace(0.1,1.1,101)
```

```python
def Function(x):
    return np.sqrt(np.tan(x))
```

```python
def DerivadaExacta(x):
    return 1/(2*(np.cos(x)**2)*(np.sqrt(np.tan(x))))

DE = DerivadaExacta(x)
print("La derivada exacta en el intervalo [0.1, 1.1] es: ")
print(DE)
```

```python
def DerivadaProgresiva(f,x,h):
    return (1/(2*h))*(-3*f(x)+4*f(x+h)-f(x+2*h))

DP = DerivadaProgresiva(Function,x,h)
print("La derivada progresiva en el intervalo [0.1, 1.1] con h = 0.01 es: ")
print(DP)
```

```python
def DerivadaCentral(f,x,h):
    
    d = 0.
    
    if h != 0:
        d = (f(x+h) - f(x-h))/(2*h)
        
    return d

DC = DerivadaCentral(Function,x,h)
print("La derivada central en el intervalo [0.1, 1.1] con h = 0.01 es: ")
print(DP)
```

```python
plt.scatter(x,DC,label='Derivada Central')
plt.scatter(x,DP,label='Derivada Progresiva')
plt.scatter(x,DE,label='Derivada Exacta')
plt.legend()
```

```python
plt.scatter(x,np.abs(DE-DP),label='Derivada Progresiva')
plt.scatter(x,np.abs(DE-DC),label='Derivada Central')
plt.yscale('log')
plt.legend()
```
