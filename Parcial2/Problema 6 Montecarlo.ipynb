```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import axes3d
from scipy import integrate
from tqdm import tqdm
N =100000
```

```python
def CreateSphere(N,R=1):
    listax = np.array([])
    listay = np.array([])
    listaz = np.array([])
    Points = np.zeros((N,3))
        
    for i in tqdm(range(N)):
            
        u = np.random.rand()
        r = R*u**(1/3)
        phi = np.random.uniform(0,2*np.pi)
        costheta = np.random.uniform(-1,1)
        theta = np.arccos((costheta))
        x = r*np.cos(phi)*np.sin(theta)
        y = r*np.sin(phi)*np.sin(theta)
        z = r*np.cos(theta)
        Points[i] = [x,y,z]
        listax = np.append(listax,x)
        listay = np.append(listay,y)
        listaz = np.append(listaz,z)
        
    return Points, listax,listay,listaz
```

```python
Points, listax,listay,listaz = CreateSphere(N)
Sphere = Points
```

```python
def I(lista1,lista2):
    suma = 0
    
    suma = np.sum(lista1**2+lista2**2)
        
    suma = suma/N
    
    return suma

Ixx = I(listay,listaz)
print("Nuestro valor calculado para el momento de inercia en Iij para i = x,y,z con i=j es: "+str(Ixx))
print("El valor teórico es: 2/5 = 0.4")
```

```python
def I2(lista1,lista2):
    suma = 0
    
    suma = np.sum(lista1*lista2)
        
    suma = suma/N
    
    return -suma
print("El producto de inercia de Ixy nos da: "+str(I2(listay,listaz)))
print("Como el producto de inercia nos da 0, significa que la figura estudiada tiene una alta simetría, por lo tanto, no tiene momentos de inercia, por lo cual, no rota.")
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(Sphere[:,0],Sphere[:,1],Sphere[:,2])
```
