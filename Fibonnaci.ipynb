```python
import numpy as np
import matplotlib.pyplot as plt
def Fibonnaci(n):
    f0 = 0
    f1= 1
    lista_respuesta = np.array([f0,f1])
    if type(n)==int:
        for i in range(n-1):
            fn = f0 + f1
            lista_respuesta = np.append(lista_respuesta,fn)
            f0 = f1
            f1 = fn
    return lista_respuesta
            
print(Fibonnaci(20))
plt.plot(Fibonnaci(20))
```

```python
numero_aureo = (1+ np.sqrt(5))/2
```

```python
def Estimacion_numero_aureo(Fibonnaci):
    numero_aureo = 0
    lista_numero_aureo = np.array([])
    for i in range(len(Fibonnaci)-1):
        fn = Fibonnaci[i]
        fn1 = Fibonnaci[i+1]
        if fn !=0:
            numero_aureo = fn1/fn
            lista_numero_aureo = np.append(lista_numero_aureo,numero_aureo)
    return lista_numero_aureo
```

```python
plt.axhline(numero_aureo,color="red", linestyle="--", label='Número aureo')
plt.plot(Estimacion_numero_aureo(Fibonnaci(20)), label='Estimación usando la serie')
```
