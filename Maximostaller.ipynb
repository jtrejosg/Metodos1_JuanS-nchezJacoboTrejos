```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
archivo = open('Data/DatosMaximo.txt', 'r')
```

```python
x = np.array([])
y = np.array([])
```

```python
for fila in archivo:
    
    fila = fila.split('  ')
    x = np.append(x,float(fila[0]))
    y = np.append(y,float(fila[1]))
    
len(y)
```

```python
plt.scatter(x,y)
```

```python
maximosx = np.array([])
maximosy = np.array([])

for i in range(0,len(y)-1):
    actual = y[i]
    anterior = y[i-1]
    siguiente = y[i+1]
    if actual > anterior and actual > siguiente:
        maximosx = np.append(maximosx,x[i])
        maximosy = np.append(maximosy,y[i])
 
```

```python
plt.scatter(maximosx,maximosy, color="red")
plt.plot(x,y)
```
