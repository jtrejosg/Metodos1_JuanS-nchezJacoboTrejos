```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funcion1(x):
    return (2*x)-2
def funcion2(x):
    return (1-x)/2
def funcion3(x):
    return 4-x

A = np.array([[2,-1],[1,2],[1,1]])
b = np.array([2,1,4])

AT = np.dot(A.T,A)
bT = np.dot(A.T,b)    
xsol = np.linalg.solve(AT,bT)
xsol
```

```python
x = np.linspace(-5,5,100)

plt.plot(x,funcion1(x))
plt.plot(x,funcion2(x))
plt.plot(x,funcion3(x))
plt.scatter(xsol[0],xsol[1],color='r')
print("Podemos ver que no existe un punto que pase por las 3 rectas ya que estas no se intersecan, pero el método de mínimos cuadrados calcula un punto que está a la menor distancia de las 3 rectas.")
```

```python
X = np.arange(-5, 5, 0.01)
Y = np.arange(-5, 5, 0.01)
X, Y = np.meshgrid(X, Y)


def funcion (x,y):
    a = 2*x-2-y
    b = x+2*y-1
    c = x+y-4
    return np.sqrt(a**2 + b**2 + c**2)
mesh = funcion(X,Y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,mesh,cmap= "RdBu_r")
ax.view_init(5,5)
ax.scatter(xsol[0],xsol[1],funcion(xsol[0],xsol[1]),color="r")
print("Graficando el valor obtenido con el método de mínimos cuadrados y la menor distancia del problema, podemos ver que la menor distancia a las 3 rectas es el punto exacto que encontramos anteriormente.")
```
