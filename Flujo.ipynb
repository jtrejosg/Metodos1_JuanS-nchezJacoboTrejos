```python
import numpy as np
import matplotlib.pyplot as plt
limite_inferior = -4
limite_superior = 4
num_puntos = 25
Radio = 2
Velocidad = 2
h = 0.001
```

```python
x = np.linspace(limite_inferior,limite_superior,num_puntos)
y = np.linspace(limite_inferior,limite_superior,num_puntos)
angulo = np.linspace(0,2*np.pi,360)
    
def Potencial(x,y):
    return Velocidad*x*(1-(Radio**2/(x**2+y**2)))

posicionx, posiciony = np.meshgrid(x,y)
posiciony
```

```python
flujo = np.nan_to_num(Potencial(posicionx,posiciony))
```

```python
Vx = (Potencial(posicionx+h,posiciony)-Potencial(posicionx-h,posiciony))/(2*h)
Vy = -1*((Potencial(posicionx,posiciony+h)-Potencial(posicionx,posiciony-h))/(2*h))
```

```python
fig = plt.figure(figsize =(5,5))
ax = fig.add_subplot()
for i in range(0,25):
    for j in range (0,25):
        norma = np.sqrt(posicionx[i,j]**2+posiciony[i,j]**2)
        if norma > 2:
            ax.quiver(posicionx[i,j],posiciony[i,j],Vx[i,j],Vy[i,j],color="blue")
            
plt.plot(Radio*np.cos(angulo),Radio*np.sin(angulo),linewidth = 3, color = "red")
```
